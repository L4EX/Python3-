方差表示的是样本点的离散程度  
那么上述问题的答案是，找到一个轴，使得样本空间的所有点映射到这个轴后，方差最大  
$Var(x) =\dfrac {1}{m} \sum^{m}_{i = 1}(x_i - \overline {x})^2 $

第一步：将样例的均值归为0(demean),每个样本减去样本均值 
$Var(x) =\dfrac {1}{m} \sum^{m}_{i = 1}(x_i)^2 $，$x_i$是原来的样本映射在新的轴上的坐标  
第二步：求新的轴的方向，w = (w1, w2),使得映射到w以后，有：$Var(X_{project}) = \dfrac {1}{m} \sum^{m}_{i = 1}(X^{(i)}_{project}- \overline X_{project})^2 $最大

 $X^{(i)}_{project}- \overline X_{project}$得到的是一个向量，故用模长表示距离  
 $Var(X_{project}) = \dfrac {1}{m} \sum^{m}_{i = 1}||X^{(i)}_{project}- \overline X_{project}||^2 $由于在新的坐标轴下，$X^{(i)}_{project}$的均值为0  
 故可以写成；$Var(X_{project}) = \dfrac {1}{m} \sum^{m}_{i = 1}||X^{(i)}_{project}||^2 $
